## Ejercicio 1

Cargad el *dataset* `employees.xlsx` de forma que obtengáis un objeto de tipo `DataFrame`.

Una vez cargado el *dataset*, resuelve las siguientes cuestiones:

1. Mostrar el número total de empleados en el DataFrame.
2. Calcular el salario máximo, mínimo y promedio de todos los empleados.
3. Mostrar cuántos empleados pertenecen a cada departamento (pista: value_counts()).
4. Crear una nueva columna Age que contenga la edad de cada empleado (a partir de *DateOfBirth*).
5. Mostrar los 5 empleados con mayor salario.
6. Mostrar cuántos empleados tienen un apellido que empieza por la letra "S".
7. Calcular la mediana de salarios de los empleados del departamento "IT".
8. Crear una nueva columna Initials que contenga las iniciales de cada empleado (ejemplo: "John Smith" → "J.S.").

**Pista 1:** Recordad que podéis utilizar librerías como `pandas` para cargar el conjunto de datos de forma sencilla y obtener un `DataFrame`.  

**Pista 2:** El *type hint* de un `DataFrame` es `pd.DataFrame`.  

**Todos los ejercicios se tienen que resolver a partir del `DataFrame` obtenido al finalizar este ejercicio y no cargarlo cada vez que se deba utilizar.**  

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 puntos)**

```python
import pandas as pd
from datetime import datetime

# Leo el archivo con ruta relativa tal como pide el enunciado.
df: pd.DataFrame = pd.read_excel("data/employees.xlsx")

# Por si acaso, fuerzo DateOfBirth a datetime (para evitar problemas si entra como string).
df["DateOfBirth"] = pd.to_datetime(df["DateOfBirth"], errors="coerce")

# Miro unas cuantas filas para confirmar que se ha leído bien.
print("vista previa del dataset:")
display(df.head())

# Calculo el número total de empleados.
total_empleados: int = len(df)
print(f"total de empleados: {total_empleados}")

# Reviso los salarios (máximo, mínimo y media).
sal_max: float = df["Salary"].max()
sal_min: float = df["Salary"].min()
sal_mean: float = df["Salary"].mean()
print(f"salario máximo: {sal_max:.2f} €")
print(f"salario mínimo: {sal_min:.2f} €")
print(f"salario medio: {sal_mean:.2f} €")

# Cuento cuántos empleados hay en cada departamento.
print("\nempleados por departamento:")
display(df["Department"].value_counts())

# Añado una columna con la edad calculada a partir de la fecha de nacimiento.
hoy = datetime.today()
df["Age"] = df["DateOfBirth"].apply(
    lambda d: hoy.year - d.year - ((hoy.month, hoy.day) < (d.month, d.day))
)

# Muestro los cinco empleados con mayor salario.
print("\nlos 5 empleados con mayor salario:")
display(df.nlargest(5, "Salary")[["FirstName", "LastName", "Salary"]])

# Cuento los empleados cuyo apellido empieza por 'S'.
n_apellidos_s: int = df["LastName"].fillna("").str.startswith("S").sum()
print(f"\nempleados con apellido que empieza por 'S': {n_apellidos_s}")

# Calculo la mediana de salario en el departamento de IT.
mediana_it: float = df.loc[df["Department"] == "IT", "Salary"].median()
print(f"\nmediana de salarios en IT: {mediana_it:.2f} €")

# Creo una columna con las iniciales (nombre y apellido).
df["Initials"] = (
    df["FirstName"].fillna("").str[0].str.upper()
    + "."
    + df["LastName"].fillna("").str[0].str.upper()
    + "."
)

# Reviso rápidamente el dataframe final.
print("\nrevisión rápida del dataframe final:")
display(df.head())
```

## Ejercicio 2

Utilizando el dataframe obtenido en el Ejercicio 1:

1. Calcular el salario medio por departamento y mostrar qué departamento tiene la media más alto.

2. Contar cuántos empleados hay en cada departamento y mostrar el resultado en orden descendente.

3. Listar los nombres y apellidos de las personas cuyo salario esté por encima del salario medio general.


<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 puntos)**

```python
# Calculo el salario medio por departamento.
salario_medio_dep = (
    df.groupby("Department")["Salary"]
    .mean()
    .sort_values(ascending=False)
)

print("salario medio por departamento:")
display(salario_medio_dep.round(2))

# Identifico el departamento con la media más alta.
dep_top = salario_medio_dep.idxmax()
media_top = salario_medio_dep.max()
print(f"\ndepartamento con salario medio más alto: {dep_top} ({media_top:.2f} €)")

# Cuento los empleados por departamento en orden descendente.
print("\nnúmero de empleados por departamento (de mayor a menor):")
empleados_por_dep = df["Department"].value_counts()
display(empleados_por_dep)

# Calculo el salario medio general.
salario_medio_general = df["Salary"].mean()

# Selecciono los empleados con salario por encima de la media general.
por_encima_media = df.loc[
    df["Salary"] > salario_medio_general,
    ["FirstName", "LastName", "Salary"],
]

print(
    f"\nempleados con salario por encima de la media general "
    f"({salario_medio_general:.2f} €):"
)
display(por_encima_media.sort_values("Salary", ascending=False))
```

Utilizando el dataframe obtenido en el Ejercicio 1:

1. Crear la función **rango_empleados** que reciba como parámetros el DataFrame y un rango de edades (min_edad y max_edad).

2. La función debe calcular la edad de cada empleado a partir de su *DateOfBirth*.

3. La función debe devolver un diccionario con:

"conteo" → número total de empleados dentro del rango de edad.

"salario_medio" → salario promedio de esos empleados.

Ejemplo:

<pre>
resultado = rango_empleados(df, min_edad=30, max_edad=50)
print(resultado)

{'conteo': 44, 'salario_medio': 53174.53}

Nota:

Se puede calcular la edad realizando la diferencia entre el año actual y el año de *DateOfBirth*

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 puntos)**

```python
from datetime import date
from typing import Dict
import pandas as pd

# Compruebo que está la columna 'Age' (si no está la creo).
if "Age" not in df.columns:
    hoy = date.today()
    df["Age"] = df["DateOfBirth"].apply(
        lambda d: hoy.year - d.year - ((hoy.month, hoy.day) < (d.month, d.day))
    )


# Defino la función rango_empleados tal como pide el enunciado.
def rango_empleados(
    df_in: pd.DataFrame, min_edad: int, max_edad: int
) -> Dict[str, float]:
    """Devuelve conteo y salario medio para empleados en [min_edad, max_edad]."""
    subset = df_in.loc[(df_in["Age"] >= min_edad) & (df_in["Age"] <= max_edad)]

    conteo: int = len(subset)
    salario_medio: float = (
        round(float(subset["Salary"].mean()), 2) if conteo > 0 else float("nan")
    )
    return {"conteo": int(conteo), "salario_medio": salario_medio}


# Pruebo la función con el rango del enunciado.
print("prueba con rango 30–50:")
resultado_demo = rango_empleados(df, min_edad=30, max_edad=50)
print(resultado_demo)

# Pruebas adicionales para comprobar que funciona bien en otros casos.
print("rango amplio (0–120):", rango_empleados(df, 0, 120))
print("rango vacío (200–300):", rango_empleados(df, 200, 300))
```

**Justificación**: En este ejercicio defino la función `rango_empleados` que recibe el DataFrame y dos valores de edad para analizar un grupo concreto de empleados.
Antes de nada, me aseguro de que la columna `Age` exista.
Dentro de la función, filtro el conjunto según las edades indicadas, cuento cuántos empleados hay y calculo su salario medio, redondeado a dos decimales.
Devuelvo el resultado en un diccionario, como pide el enunciado.
Por último, pruebo la función con tres rangos diferentes para comprobar que gestiona bien tanto casos normales como extremos (ej: un rango sin empleados).

## Ejercicio 4

1. Dada una lista de números de teléfono, debéis recorrerla y, aplicando expresiones regulares, determinar si el número es válido o no válido.

Un número válido debe tener el formato:

- Opcional +34 o 0034 al inicio (código de país).
- Seguido de 9 dígitos.
- Puede tener espacios, guiones o puntos como separadores entre los dígitos.
- Ejemplos válidos: +34 612-345-678, 0034.612.345.678, 612345678.

2. Dado un texto que contiene horas, debéis extraer, a través de una única expresión regular, una lista con todas las horas que tengan el formato correcto:

- Formato correcto: HH:MM o HH:MM:SS
- HH: 0–23
- MM y SS: 0–59
- Ejemplo válido: 08:30, 23:59:59, 0:5:7 (permitiendo 1 o 2 dígitos).

Podéis utilizar para validar el ejercicio los siguientes ejemplos de uso:

<pre>
# Lista de números de teléfono
numero_movil = [
    "+34 612-345-678",
    "0034.612.345.678",
    "612345678",
    "+34 612 345 67",    # inválido
    "12345678",           # inválido
    "+341234567890"       # inválido
]

# Extracción de horas

texto_con_tiempos = """

</pre>
<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 puntos)**

```python
import re
from typing import List

# a) Valido números de teléfono, siguiendo el enunciado.
pat_tel = re.compile(r"^(?:\+34|0034)?[ .-]?\d{3}[ .-]?\d{3}[ .-]?\d{3}$")

numeros_prueba: List[str] = [
    "+34 612-345-678",
    "0034.612.345.678",
    "612345678",
    "+34 612 345 67",  # inválido
    "12345678",        # inválido
    "+341234567890",   # inválido
]

# Separo válidos e inválidos con list comprehensions.
validos = [n for n in numeros_prueba if pat_tel.fullmatch(n)]
invalidos = [n for n in numeros_prueba if not pat_tel.fullmatch(n)]

print("teléfonos válidos:")
print(validos)
print("\nteléfonos inválidos:")
print(invalidos)

# b) Extraigo horas con formato HH:MM o HH:MM:SS permitiendo uno o dos dígitos por bloque.
pat_hora = re.compile(r"\b(?:[01]?\d|2[0-3]):(?:[0-5]?\d)(?::(?:[0-5]?\d))?\b")

texto_con_tiempos = (
    "La reunión empieza a las 08:30 y termina a las 17:45:30.\n"
    "Otra hora errónea: 25:61, y otra correcta: 0:5:7.\n"
)

# Extraigo todas las coincidencias válidas.
horas = pat_hora.findall(texto_con_tiempos)
print("\nhoras válidas encontradas en el texto:")
print(horas)
```

**Justificación**: En este ejercicio he trabajado con expresiones regulares para validar formatos de texto.
Primero, defino un patrón que reconoce números de teléfono españoles, con la opción de incluir el prefijo +34 o 0034 y separadores como espacios, guiones o puntos.
Decido usar list comprehensions para separar válidos e inválidos porque son más compactas y fáciles de leer que un bucle `for`.
Después, creo otro patrón para detectar horas válidas con los formatos `HH:MM` o `HH:MM:SS`, permitiendo también que los valores tengan un solo dígito.
Por último, pruebo el patrón con un texto de ejemplo que incluye tanto horas válidas como incorrectas, para asegurar que la expresión cumple con las restricciones de rango (0–23 para horas, 0–59 para minutos y segundos).

## Ejercicio 5

Utilizando el dataframe obtenido en el Ejercicio 1:

1. Crea una nueva columna llamada **Categoria_Trabajo** que contenga la categoría del trabajo según la columna *JobTitle*.

Por ejemplo:
- Si *JobTitle* contiene "Engineer" → "Engineering"
- Si *JobTitle* contiene "Manager" → "Management"
- Si *JobTitle* contiene "Teacher" → "Education"
- Si no cumple ninguna de las anteriores → "Other"

2. Crea un diccionario **category_counts** que contenga el número de empleados por cada *JobCategory* usando dict comprehension.

3. Ordena el diccionario por número de empleados descendente y guárdalo como *OrderedDict*.

4. Imprime el diccionario recorriéndolo con un bucle for.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px;">SM</span> **(2 puntos)**

```python
from collections import OrderedDict
from typing import Dict

# Defino una función para clasificar los trabajos según la palabra clave del título.
def asignar_categoria(job: str) -> str:
    j = (job or "").lower()
    if "engineer" in j:
        return "Engineering"
    if "manager" in j:
        return "Management"
    if "teacher" in j:
        return "Education"
    return "Other"


# Aplico la función a la columna 'JobTitle' para crear una nueva categoría más general.
df["Categoria_Trabajo"] = df["JobTitle"].apply(asignar_categoria)

# Calculo cuántos empleados hay en cada categoría usando una dict comprehension.
category_counts: Dict[str, int] = {
    cat: int((df["Categoria_Trabajo"] == cat).sum())
    for cat in df["Categoria_Trabajo"].unique()
}

# Ordeno el diccionario de forma descendente por número de empleados.
category_counts = OrderedDict(
    sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
)

# Recorro el diccionario e imprimo los resultados.
print("número de empleados por categoría de trabajo:")
for categoria, cantidad in category_counts.items():
    print(f"{categoria}: {cantidad}")

# Muestro algunos ejemplos de cómo ha quedado la nueva columna.
print("\nejemplo de la nueva columna 'Categoria_Trabajo':")
display(df[["JobTitle", "Categoria_Trabajo"]].head())
```

**Justificación**: En este ejercicio agrupo los distintos puestos de trabajo en categorías generales para simplificar el análisis.
Primero, defino una función `asignar_categoria()` que revisa la columna `JobTitle` y asigna una categoría según las palabras clave más comunes (`Engineer`, `Manager`, `Teacher`).
Si un título no coincide con ninguno de esos casos, lo clasifico como "Other", ya que siempre conviene tener una categoría por defecto.
Después, utilizo una dict comprehension para contar el número de empleados por categoría.
Finalmente, ordeno el resultado con `OrderedDict` para mostrar las categorías desde la más frecuente a la menos frecuente y muestro un ejemplo de cómo se añade la nueva columna al DataFrame.

**Referencia:** Unidades 0 y 1 de la asignatura.
